```python
import pandas as pd
from nltk.corpus import stopwords
from textblob import TextBlob
import re
from textblob import Word
```

```python
data=pd.read_csv('Mac_0402_TextAnalysis.csv')
data
```

```python
data['word_count'] = data['text'].apply(lambda x: len(str(x).split(" ")))
```

#### 2. Preprocessing

##### 2.1 Lower Case

```python
data['text'] = data['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

##### 2.2 Removing Punctuations

```python
data['text'] = data['text'].str.replace('[^\w\s]','')
```

###### 2.3 Removing Stop Words

```python
stop = stopwords.words('english')
data['text'] = data['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop ))
```

###### 2.4 Words removal

As the data all related to Mac Cosmetics, here we'd love to identify the common words which we'd love to remove.

```python
freq = pd.Series(' '.join(data['text']).split()).value_counts()[:20]
freq
```

```python
freq_top=pd.Series(' '.join(data['text']).split()).value_counts()[:10]

freq_top=list(freq_top.index)
```

```python
data['text'] = data['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_top ))
```

```python
#freq_bottom=pd.Series(' '.join(data['text']).split()).value_counts()[-10:]
#freq_bottom
```

##### 2.6 Link Removal

```python
data['text']=data['text'].apply(lambda x: ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", x).split()))
```

```python
data['text'].head()
```

```python
data['text_nolink']=data['text'].apply(lambda x: re.sub(r"http\S+", "", x))
```

```python
data['text_nolink'].head()
```

###### 2.6 Tokenization

```python
data['token']=data['text_nolink'].apply(lambda x: TextBlob(x).words)
```

```python
data['token'].head()
```

###### 2.7  Lemmatization - Root words

```python
data['lemmatization'] = data['text_nolink'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data['lemmatization'].head()
```
